```python
df['gender'].replace(['Male','Female'],value=[1,0],inplace=True)
df['Residence_type'].replace(['Urban','Rural'],value=[1,0],inplace=True)
```

```python
df['bmi'].fillna(df['bmi'].mean(),inplace=True)
```

```python
df['ever_married'].replace(['Yes','No'],value=[1,0],inplace=True)
```

```python
df[pd.get_dummies(df['work_type'],drop_first=True).columns]=pd.get_dummies(df['work_type'],drop_first=True)
df[pd.get_dummies(df['smoking_status'],drop_first=True).columns]=pd.get_dummies(df['smoking_status'],drop_first=True)
```

```python
df=df.drop(['Residence_type','Private','smokes','id','never smoked','Never_worked','children'],axis=1)
```

```python
df.drop(['work_type','smoking_status'],inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
sam=df.sample(random_state=101,n=1250)
```

```python
x=sam.drop(['Self-employed', 'formerly smoked','stroke'],axis=1)
y=sam['stroke']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=101,test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
```

```python
lr.fit(xtrain,ytrain)
pred=lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(ytest,pred))
```
